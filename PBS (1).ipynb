```python
class HuffmanCoding:
    def __init__(self, path):
        self.path = path
        self.heap = []
        self.codes = {}
        self.reverse_mapping = {}
        
    class HeapNode:
        def __init__(self, char, freq):
            self.char = char
            self.freq = freq
            self.left1 = None
            self.left2 = None
            self.right1 = None
            self.right2 = None
        
        def __lt__(self, other):
            return self.freq <= other.freq

        def __eq__(self, other):
            if(other == None):
                return False
            if(not isinstance(other, HeapNode)):
                return False
            return self.freq == other.freq
   
        
    def make_probability_dict(self, prob, index):
        probability = {}
        #print(prob)
        for i in range(len(prob)):
            probability[index[i]] = prob[i]
            #print(index[i], probability[index[i]])
        return probability

    def make_heap(self, probability):
        #print(probability)
        for key in probability:
            #print(key)
            node = self.HeapNode(key, probability[key])
            heapq.heappush(self.heap, node)
        
    def merge_nodes(self):
        while(len(self.heap)>1):
            node1 = heapq.heappop(self.heap)
            node2 = heapq.heappop(self.heap)
            node3 = heapq.heappop(self.heap)
            node4 = heapq.heappop(self.heap)
            merged = self.HeapNode(None, node1.freq + node2.freq + node3.freq + node4.freq)
            merged.left1 = node1
            merged.left2 = node2
            merged.right1 = node3
            merged.right2 = node4
            heapq.heappush(self.heap, merged)
        
    def make_codes_helper(self, root, current_code):
        if(root == None):
            return

        if(root.char != None):
            self.codes[root.char] = current_code
            self.reverse_mapping[current_code] = root.char
            return

        self.make_codes_helper(root.left1, current_code + "0")
        self.make_codes_helper(root.left2, current_code + "1")
        self.make_codes_helper(root.right1, current_code + "2")
        self.make_codes_helper(root.right2, current_code + "3")

    def make_codes(self):
        root = heapq.heappop(self.heap)
        current_code = ""
        self.make_codes_helper(root, current_code)
```

```python
h = HuffmanCoding(None)
#print(prob)
#print(index)
probability = h.make_probability_dict(prob, index)
#print(probability)
h.make_heap(probability)
h.merge_nodes()
h.make_codes()
print(h.codes)
```
